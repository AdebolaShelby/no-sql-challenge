5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

# Update all documents where longitude and latitude are stored as strings

documents = establishments.find({"geocode.longitude": {"$type": "string"}, "geocode.latitude": {"$type": "string"}})

# loop through the documents and update the data type

for doc in documents:
    longitude = Decimal128(doc["geocode"]["longitude"])
    latitude = Decimal128(doc["geocode"]["latitude"])
    establishments.update_one({"_id": doc["_id"]}, {"$set": {"geocode.longitude": longitude, "geocode.latitude": latitude}})

    # Update the document with the converted values

    establishments.update_one({"_id": doc["_id"]}, {"$set": {"geocode.longitude": longitude, "geocode.latitude": latitude}})


# Check that the data type has been updated
query = ["geocode.longitude", "geocode.latitude"]
result = establishments.find_one({}, query)
pprint(result)

for key, value in result.items():
    print(f"The data type for {key} is {type(value)}")
    
```

```python
# Check that the data type was updated
query = ["geocode.longitude", "geocode.latitude"]
result = establishments.find_one({}, query)
pprint(result)

for key, value in result.items():
    print(f"The data type for {key} is {type(value)}")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue

# find documents where RatingValue is stored as a string
documents = establishments.find({"RatingValue": {"$type": "string"}})

# loop through the documents and update the data type

for doc in documents:

    try:
        # convert the string to an integer
        rating = int(doc["RatingValue"])
        establishments.update_one({"_id": doc["_id"]}, {"$set": {"RatingValue": rating}})

    except ValueError:
        # if the string cannot be converted to an integer, set the value to None
        establishments.update_one({"_id": doc["_id"]}, {"$set": {"RatingValue": None}})

# Check that the data type has been updated
query = ["RatingValue"]
result = establishments.find_one({}, query)
pprint(result)

for key, value in result.items():
    print(f"The data type for {key} is {type(value)}")
    
```

```python
# check that the coordinates and rating value are now numbers
query = ["geocode.longitude", "geocode.latitude", "RatingValue"]
result = establishments.find_one({}, query)

for key, value in result.items():
    print(f"The data type for {key} is {type(value)}")
```
